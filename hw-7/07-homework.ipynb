Сделаем куски датасета, чтобы передать их как train и test

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("../data/singapore_airlines_reviews.csv")
df_09, df_01 = train_test_split(df, test_size=0.1, random_state=42)
df_09.to_csv("singapore_airlines_reviews_09.csv")
df_01.to_csv("singapore_airlines_reviews_01.csv")
```

```python
! python3 main.py train --data "singapore_airlines_reviews_09.csv" --test "singapore_airlines_reviews_01.csv" --model model.pkl
```
